```python
import base64

def convert_mp3_to_base64(file_path):
    # Open the MP3 file in binary read mode
    with open(file_path, 'rb') as mp3_file:
        # Read the entire file
        mp3_data = mp3_file.read()
        # Encode the binary data to Base64
        base64_encoded_data = base64.b64encode(mp3_data)
        # Convert the Base64 bytes to a string
        base64_string = base64_encoded_data.decode('utf-8')
    return base64_string

# Specify the path to your MP3 file
file_path = "C:\\Users\\Anand\\Downloads\\hindi.mp3"

# Convert the MP3 file to a Base64 string
base64_string = convert_mp3_to_base64(file_path)

# Optional: Save the Base64 string to a file (if needed)
with open('output_base64.txt', 'w') as file:
    file.write(base64_string)

print("MP3 file has been converted to Base64 and saved to output_base64.txt")
```
